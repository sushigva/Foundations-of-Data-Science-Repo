## Pie Chart (3 points)

```python
# Create a pie chart from raw values

slices = [56, 66, 24, 124, 47]
labels = ["US","UK","MX","CA","JP"]
```

```python


import matplotlib.pyplot as plt

# these are the raw values from above
slices = [56, 66, 24, 124, 47]
labels = ["US", "UK", "MX", "CA", "JP"]

# this is the cration of the pie chart
plt.pie(slices, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Pie chart from raw values')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# the random data
n = 430 + np.random.randn(3000)

# making histogram
plt.hist(n, bins=30, color='black', alpha=0.2)
plt.xlabel('Data Values')
plt.ylabel('Frequency')
plt.title('The histogram of random data')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# above
num_points = 50
x = 5 + np.arange(num_points) + np.random.randn(num_points)
y = 10 + np.arange(num_points) + 5 * np.random.randn(num_points)

# fitting a linear trendline
coefficients = np.polyfit(x, y, 1)
trendline = np.polyval(coefficients, x)

# making scatterplot
plt.scatter(x, y, label='the Data Points')

# Plotting the trendline, chose yellow
plt.plot(x, trendline, color='yellow', label='the Trendline')

# legned/labels
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatterplot with Trendline from two lists of values')
plt.legend()

# shows it
plt.show()
```
